```python
data = pd.read_csv('./predict-vs-real/2020/linha41_1mes(3)_1semanas(4)_predict.csv', delimiter=',')


real_values = data.validations_per_hour
predictions = data.validations_per_hour_predicted

plt.figure(figsize=(10,10))
plt.scatter(real_values, predictions, c='crimson')

p1 = max(max(predictions), max(real_values.values))
p2 = min(min(predictions), min(real_values.values))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.title('Linha 041: Treinamento mes completo de Março -> Previsao da primeira semana de Abril - 2020')
plt.xlabel('Valores Reais', fontsize=15)
plt.ylabel('Previsoes', fontsize=15)
plt.axis('equal')
# plt.savefig('./imagens/2020/sazonalidade_hora.pdf')
plt.show()

data = pd.read_csv('./predict-vs-real/2018/linha41_1mes(3)_1semanas(4)_predict.csv', delimiter=',')

real_values = data.validations_per_hour
predictions = data.validations_per_hour_predicted

plt.figure(figsize=(10,10))
plt.scatter(real_values, predictions, c='crimson')

p1 = max(max(predictions), max(real_values.values))
p2 = min(min(predictions), min(real_values.values))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.title('Linha 041: Treinamento mes completo de Março -> Previsao da primeira semana de Abril - 2018')
plt.xlabel('Valores Reais', fontsize=15)
plt.ylabel('Previsoes', fontsize=15)
plt.axis('equal')
plt.show()

data = pd.read_csv('./predict-vs-real/2015/linha41_1mes(3)_1semanas(4)_predict.csv', delimiter=',')

# data
real_values = data.validations_per_hour
predictions = data.validations_per_hour_predicted


plt.figure(figsize=(10,10))
plt.scatter(real_values, predictions, c='crimson')

p1 = max(max(predictions), max(real_values.values))
p2 = min(min(predictions), min(real_values.values))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.title('Linha 041: Treinamento mes completo de Março -> Previsao da primeira semana de Abril - 2015')
plt.xlabel('Valores Reais', fontsize=15)
plt.ylabel('Previsoes', fontsize=15)
plt.axis('equal')
plt.show()
```
