```python
import pandas as pd
import glob
import matplotlib.pyplot as plt

# All files ending with .txt
performances_file_list = glob.glob("./performances/*.csv")
predict_vs_real_file_list = glob.glob("./predict-vs-real/*.csv")
print(performances_file_list)
print(predict_vs_real_file_list)
```

```python
for i in range(3, 13):
    data = pd.read_csv(f'./predict-vs-real/linha41_1mes({i})_2semanas(4)_predict.csv', delimiter=',')
    real_values = data.validations_per_hour
    predictions = data.validations_per_hour_predicted
    p1 = max(max(predictions), max(real_values.values))
    p2 = min(min(predictions), min(real_values.values))
    plt.figure(figsize=(10,10))
    plt.scatter(real_values, predictions, c='crimson')
    plt.plot([p1, p2], [p1, p2], 'b-')
    plt.xlabel('True Values', fontsize=15)
    plt.ylabel('Predictions', fontsize=15)
    plt.title(f'MES {i} PREVENDO MES 4')
    plt.axis('equal')
    plt.show()
```
